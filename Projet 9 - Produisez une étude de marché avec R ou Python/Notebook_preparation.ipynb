# Création d'une table par une concaténation entre les dataframes 

Je vais réalisé une concaténation des dataframes, ce qui consiste à ajouté les dataframe les uns en dessous des autres, pour centraliser les données et réaliser des analyses dessus.

```python
data = pd.concat([dispoAlimen,pop_20_22,stabpol_20_22])
data
```

J'affiche les dimention du dataframe ainsi que les différentse informations sur celui-ci.

```python
print("La table comporte", data.shape[0], "lignes et", data.shape[1], "colonnes.")
```

```python
data.info()
```

```python
print("Il y a", data["Zone"].nunique(), 'pays différents dans le fichier centralisé.')
```

Pour une meilleur lisibilité des données, je pivote la colonne pays afin d'avoir toute les données sur 1 seul ligne pour chaque pays

```python
data = data.pivot_table(index='Zone', values='Valeur', columns=['Élément', 'Produit', 'Année']).reset_index()
data
```

```python
exports, imports, kcal, prot, pop, pop1, poulet, production, prix_prod, pib, politique, secu = 'Exportations - quantité', 'Importations - quantité', 'Disponibilité alimentaire (Kcal/personne/jour)', 'Disponibilité de protéines en quantité (g/personne/jour)', 'Population totale', 'Population-Estimations', 'Viande de Volailles' , 'Production', 'Prix à la Production (USD/tonne)', 'PIB par habitant, ($ PPA internationaux constants de 2017)', 'Stabilité politique et absence de violence/terrorisme (indice)', 'Valeur'

annee_1, annee_2 = 2000, 2022

df = pd.DataFrame()

df['pays'] = data['Zone']
df['dispo_calories'] = data[kcal][poulet][annee_2]
df['dispo_proteines'] = data[prot][poulet][annee_2]
df['population'] = data[pop][pop1][annee_2]
df['pop_croissance'] = data[pop][pop1][annee_2] / data[pop][pop1][annee_1]
df['pib'] = data[secu][pib][annee_2]
df['pib_croissance'] = data[secu][pib][annee_2] / data[secu][pib][annee_2]
df['politique'] = data[secu][politique][annee_2]
df['importation'] = data[imports][poulet][annee_2]


df
```

##### Gestion des données manquantes

```python
df = df.replace([np.inf, -np.inf], np.nan).dropna().reset_index().drop(['index'], axis=1)
df
```
